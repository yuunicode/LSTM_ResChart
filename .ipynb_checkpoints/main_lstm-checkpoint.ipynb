```python
import random
import numpy as np
from module.generate_data import Data
from module.lstm import Optimization, LSTMModel
from torch.utils.data import TensorDataset, DataLoader
import torch.nn as nn
import torch.optim as optim
import torch
from sklearn.preprocessing import StandardScaler
import pickle
```

```python
data = Data()
random.seed(421)
x1 = data.ar_1(size=1800, param=0.5)
random.seed(4242)
x2 = data.ar_1(size=600, param=0.5)
random.seed(424242)
x3 = data.ar_1(size=600, param=0.5)
```

```python
scaler = StandardScaler()

train_x, train_y = data.polish_up(x1)
vali_x, vali_y = data.polish_up(x2)
test_x, test_y = data.polish_up(x3)

scaler.fit(train_x)
train_x = scaler.transform(train_x)
vali_x = scaler.transform(vali_x)
test_x = scaler.transform(test_x)

train_x = torch.Tensor(train_x)
vali_x = torch.Tensor(vali_x)
test_x = torch.Tensor(test_x)

print(train_y.shape)

scaler.fit(train_y.reshape(-1,1))
train_y = scaler.transform(train_y.reshape(-1,1))
vali_y = scaler.transform(vali_y.reshape(-1,1))
test_y = scaler.transform(test_y.reshape(-1,1))

train_y = torch.Tensor(train_y.flatten())
vali_y = torch.Tensor(vali_y.flatten())
test_y = torch.Tensor(test_y.flatten())
```

```python
train = TensorDataset(train_x, train_y)
val = TensorDataset(vali_x, vali_y)
test = TensorDataset(test_x, test_y)

batch_size = 150
train_loader = DataLoader(train, batch_size, shuffle=False)
val_loader = DataLoader(val, batch_size, shuffle=False)
test_loader = DataLoader(test, batch_size, shuffle=False)
```

```python
input_size = 24
hidden_size = 24
num_layers = 1
output_size = 1
```

```python
lstm = LSTMModel(input_size, hidden_size, num_layers, output_size)
```

```python
learning_rate = 1e-5
epochs = 300
loss_fn = nn.MSELoss()
optimizer = optim.Adam(lstm.parameters(), lr=learning_rate, weight_decay=0.1)
# scheduler = optim.lr_scheduler.LambdaLR(optimizer=optimizer, lr_lambda = lambda epoch : 0.95**epoch)
```

```python
opt = Optimization(model=lstm, loss_fn=loss_fn, optimizer=optimizer)
# opt = Optimization(model=lstm, loss_fn=loss_fn, optimizer=optimizer, scheduler=scheduler)
opt.train(train_loader, val_loader, epochs)
```

```python
opt.plot_losses()
```

```python
model_path = './weights/'
torch.save(lstm.state_dict(), model_path+'model_state_dict')
```

```python
predictions, values = opt.evaluate(test_loader)
```

```python
len(predictions)
```

```python
pred = []
for sublist in predictions:
    for subsub in sublist:
        for item in subsub:
            pred.append(item)
len(pred)
```

```python
true = []
for sublist in values:
    for subsub in sublist:
        true.append(subsub)
len(true)
```

```python
np.mean(pred), np.std(pred)
```

```python
error = np.subtract(true, pred)
ebar = np.mean(error)
ebar, np.mean(np.square(error))
```

```python
sum = 0
for i in range(0, len(error)-1):
    sum += abs(error[i+1] - error[i])
mrbar = sum/(len(error)-1)
```

```python
l = 4.43
ucl = ebar + (l*mrbar/1.128)
lcl = ebar - (l*mrbar/1.128)
ucl, lcl
```

```python
from tqdm import tqdm

count = 1000
rl = 0

for t in tqdm(range(0, count)):
    
    data = Data()
    X = data.ar_1(1000, 0.5)
    x, y = data.polish_up(X)
    scaler_x = StandardScaler()
    x = scaler_x.fit_transform(x)
    x = torch.Tensor(x)
    scaler_y = StandardScaler()
    y = scaler_y.fit_transform(y.reshape(-1, 1))
    y = torch.Tensor(y)
    dataset = TensorDataset(x, y)
    data_loader = DataLoader(dataset, batch_size, shuffle=False)
    prediction, value = opt.evaluate(data_loader)

    yhat = []
    for sublist in prediction:
        for subsub in sublist:
            for item in subsub:
                yhat.append(item)
    target = []
    for sublist in value:
        for subsub in sublist:
            target.append(subsub)
    res = np.subtract(yhat, target)
    for i in range(0, len(res)):
        if (res[i] > ucl).any():
            break
        elif (res[i] < lcl).any():
            break
    rl += i    
arl = rl/count
arl
```
